# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks=int(input("Enter your marks: "))
if marks >= 90:
    print(f"Your grade is: A with {marks}%")
    
elif marks >= 80:
    print(f"Your grade is: B with {marks}%")

elif marks >= 70:
    print(f"Your grade is: C with {marks}%")
    
elif marks >= 60:
    print(f"Your grade is: C with {marks}%")

else:print(f"You are fail: with {marks}%")    
        
    
        
    
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
months=int(input("Enter month number:"))
if months == 12 or months == 1 or months==2:
    print("Winter")
elif months >= 3 and months<6:
    print("Spring")
elif months >= 6 and months<9:
    print("Summer")
else:print("Autumn")    
    
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = int(input("Enter num1: "))
opr = input("Enter operator (+, -, *, /): ") 
num2 = int(input("Enter num2: "))
if opr == "+":
    print(num1 + num2)
elif opr == "-":
    print(num1 - num2)    
elif opr == "*":
    print(num1 * num2)       
elif opr == "/":
    print(num1 / num2) 
else:
    print("Invalid operator")
```
